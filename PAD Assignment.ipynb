### Random generator:

The random generator serves as a replacement for *RandomState*. The difference with RandomState(legacy) is that the generator relies on an additional BitGenerator to manage state and draw the the random bits that can be then turn into random values from useful distributions([source](https://numpy.org/doc/stable/reference/random/generator.html#simple-random-data)) while RandomState uses the legacy seeding algorithm.

### Simple random data functions :

 Simple random data functions generate arbitrary values, however you are able to add requirements to this generated lists of values. For example 'randint' function returns arbitrary integers from the lowest integer to highest.The optional parameter 'dtype' corresponds to the data type: 'numpy.random.randint(low, high=None, size=None, dtype)'.

Inside of this category we have 4 functions which I will explain then in more detail below:


#### Numpy.random.Generator.integers: 

Returns random integers from low (inclusive) to high (exclusive) or both inclusive if endpoint = True. These integers come from the "discrete uniform" of the data type specified. 

In a discrete uniform distribution an event has the same chances of happening. 




See example below of the use and representation of this function. 

```python
#In order to observe the discrete uniform distribution we want to achieve with this function. We need to plot a 
#bigger array of integers: 

rng = np.random.default_rng() #we enter this line of code to create a new generator with default (PCG64)
rng.integers(200, size=10000)

x= rng.integers(200, size=100000)

print(x)

#Plotting this array with bigger size using an histogram we can observe the uniform distribution which is similar to a
#rectangle as it means that all numbers have the same probablitiy of occuring 

plt.hist(x)

plt.title('Outputs using generator integers')

plt.show()
```

#### Numpy.random.Generator.random: 

This function returns random floats considering a half-open interval .The floats generated are result from the continuos uniform distribution over the stated interval. 
                                                                      

Example below using the 'random randint function' which uses a discrete uniform distribution .Any of the 1 to 6 possible is equally possible to happen.

```python
#In this example I generate a an array of integers between 0 and 7 exclusive

np.random.randint (1,7,6)

x= np.random.randint (1,7,6)

print(x)

#Plotting this random sample

plt.hist(x)

plt.title('Outputs using generator random')

plt.show()
```

#### Numpy.random.Generator.choice: 

This one generates a random sample from a given 1 dimension array. 
The parameters are 'a' which is a 1-D array or an integer, 'size','replace' and 'p' which is the probability linked to the entries in 'a'. These last 3 parameters are optional and if not entered the value returned will be a single value , no replacement by default and with an uniform distribution if not chosen any for 'p'. 


```python
#Example of this function usage

#We create an array of 12 integers between 1 and 9 without replacement and uniform distribution by default 
np.random.choice(a= np.arange(1,30),size = 12, replace = False, p = None)
```

```python
#Plotting the integers above to show distribution more clearly: 
x = np.random.choice(a= np.arange(1,25),size = 19, replace = False, p = None)

print(x)

plt.hist(x)

plt.title('Outputs using generator choice')

plt.show()
```

We can observe with the plot above that the sample integers are distributed in a way that resembles a rectangle which is characteristic of an uniform distribution. 
On the other hand if we want to create a random sample that does not follow an uniform distribution we can add a probability value using 'p' parameters.



The sequence can be a string, a range, a list or a tuple. Let's see an example using a list:

```python
#We create a list  for this example: 

samplelist = ['Name','ID','Department','Address']

final_list= rng.choice(samplelist,4,p=[0.15,0.35,0.05,0.45])#We add  the list,size and probability for every outcome

print(final_list)

#Now we use a simple histogram to show the elements of the list 
#and how the one with the higher probability shows more often.    

plt.hist(final_list)

plt.show()
```

### Permutation functions 

#### Numpy.random.Generator.shuffle:

Modifies a sequence by shuffling its elements. Let's see an example in how to use this function below: 

```python
# We add parameter size to the function and see how the method reorganize the order of the items
#but this function does not generate a new list. 

arr = ([3,6,9,11,15])#Here we specify we want an array of integers in a  range of 0 to 15 (inclusive)
np.random.shuffle(arr)# The function then shuffles the contents of this array

arr
```

#### Numpy.random.Generator.permutation:

### 1.Chisquare distribution:
 
 This distribution is usually used in hypothesis testing. It is one method to show a relationship between two categorical variables and tells you how much difference between a count you observed ('O' value in formula below) and the counts you would expect ('E' value as seen on formula below) when there is no relationship in the population.  

 



Numpy chisquare function draw samples from a chi-square distribution. 
The parameters this function use are 'df' for the degree of freedom which are the values that can vary from the rest in the data sample and 'size' of the array we will get. 




```python
#Example of chisquare distribution function:

import seaborn as sns

x= np.random.default_rng().chisquare(df=3, size=(1,3000)) #Entered array size and 3 as degree of freedom

print(x)

#Plotting the result of the random chisquare distribution function:

sns.distplot(x)
plt.title('Chisquare distribution')
plt.show()
```

### 2.Exponential distribution:

The exponential distribution is usually used to indicate the time elapsed between events. One of the main features of this type of distribution is that it is memoryless, as the past events will not influence future events. For example if you are tossing a few coins until they show heads and do not check if there are heads, it would be like we have started all over again. 
The events ( in this case the coin tossing) are independent. 



The numpy.random.exponential function draw samples from an exponential distribution.



```python
import matplotlib.pyplot as plt

x = np.random.exponential(5,30)
print(x)

#Here there is an example of plotting the exponential distribution with a histogram and a distplot 

ax= sns.distplot(x,color='orange')

plt.title('Exponential distribution')
plt.show()
```

We can observe that the plots are showing the usual shape of a exponential distribution.

### 3.Gamma distribution :

We use this function to draw samples from Gamma distribution. Gamma distribution has the parameters: 'shape' (also called 'k') and 'scale'(also known as 'theta') and both must be greater than 0. 

This distribution differs from the 'exponential distribution' since the last one predicts what is the wait time until the first event happens while gamma distribution predicts the wait time until the event specified ( the event we have indicated with parameters 'k' and 'theta') happens. 

The Gamma distribution is often use to model positive continuous variables. Distributions are often skewed as in this case waiting times between events are relevant. 



```python
#In this example we consider shape and scale and use an histogram along
#with the probability density function to show how the floats are spreading on this distribution:

import scipy.special as sps

shape, scale = 5,3
s = np.random.gamma(shape,scale,3000)
print(s) 

#We plot this distrbution 

count, bins, ignored = plt.hist(s, 50, density=True)

y = bins**(shape-1)*(np.exp(-bins/scale) /  
                     (sps.gamma(shape)*scale**shape))
plt.plot(bins, y, linewidth=2, color='b')  
plt.title('Gamma distribution')
plt.show()
```

We can observe that the graphic shows a right skewed distribution.

### 4.Binomial distribution:

The binomial distribution finds the probability of a number of successes in a specific number of trials .It counts only two possible outcomes , usually considered as 1(success) and 0(failure).
This type of distribution is often used for cases in where there are two possible outcomes such as Republican or Democrat winning an election or head or tails when coin flipping. 

The random binomial distribution draw samples from this type of distribution.The parameters used are 'n' as an integer or array (>= 0), 'p' as an integer (>= 0 and <=1) and size. 

Let's see an example below in which we can see how to use this function and create a plot with its output. 



```python
#Let's think about a coin being flipped 20 times, using 10 as size

rng = np.random.default_rng()
n,p= 20,0.5 #here we add the number of trials'n' and probability of each trial 'P' this value between 0 and 1
sample = rng.binomial(n,p,500)

print(sample)

#Plotting this example: 
count, bins, ignored = plt.hist(sample, 10, density=True, facecolor='red', alpha=0.77)
plt.title('Binomial distribution') 
plt.ylabel('p(Probability of success)')
plt.xlabel('Trials') 
plt.show()
```

### 5. Poisson distribution:

This function provides samples from a Poisson distribution. The Poisson distribution can be used to estimate the repetition of an event in a fixed time interval or region of opportunity. The two parameters used are 'lam'() this is the number of events or occurences and 'size' of the array we obtain.Let's see examples below:

Generally speaking computers are not able to generate random numbers. However is a mechanism that helps to make computers to generate what it looks like a random number, this mechanism is called 'seeding'.
The seed is the position in which we start before generating random numbers, we can guess what random number the computer is going to generate if we know the seed. 

You can use random generator functions in numpy in order to select your seed. 
But how does the random seed module work?

 First ,using a function as random.seed () you enter  a value for example 'random.seed(3)' and then this number will be like the starting point for the samples we obtain. You can also use strings, bytes and bytearrays which will be converted into integers in order to start the seeding process. 
 
 This method provides the same random number generation so it can be used for testing. However this method of pseudo random number generator is not reliable in terms of security due to the following reasons: 
 
 -The random number needs to be cryptographically secure which is determined by entropy and size. 
 1. *Entropy* means that here is some factor that makes the number difficult to be guessed.
 2. *Size* is an importanto factor as the larger the number the more difficult could be for someone to deduce the pattern.

#### Differences between Mersenne Twister and PCG-64 generator:


In this section I am detailing how each of the algorithms mentioned above work and how PCG-64 it is doing better than the Mersenne Twister algorithm.




##### How the Mersenne Twister algorithm works : 

This algorithm was created in 1997 by Matsumoto & Nishimura and it 623-dimensionally equidistributed which means that every 623-tuple will occur once. 

The way it works is like another pseudo random number generators , it starts with a seed initialize into a state . Then when you want to draw a random number the state is transformed with one -way function ( in order to keep security so people don't find out the state from the output).In the case of Mersenne Twister the first state is not used to output the random numbers and the state with MT allow you to output 624 random numbers instead of only 1. The g function(called "temper") in the case of Mersenne Twister is reversible , which makes this algorithm not cryptographically safe.In order to create another random value  you use the function f(called "twist")  to transform the state. 

This algorithm produces 32 or 64-bit numbers which makes it usable as source of random bits


 ##### How the PCG-64 algorithm works:
 
 The PCG-64 or Permuted Congruential Generator was developed in 2014 by Melissa E. O'Neill.
 It is part of a family of simpler ,faster and statistically good algorithms to be used for random number generation. 
 This algorithm needs a seed to initialize the BitGenerator. In case none, then unpredictable entropy will be pulled from the OS, while if an int or array is passed it pass to a SeedSequence to obtain the initial BitGenerator state [(source)](https://numpy.org/doc/stable/reference/random/bit_generators/pcg64.html).

 
 
 
 
 

Some of the advantages of PCG-64 over Mersenne Twister are  :
 
 - The PCG-64 is more compact in terms of space usage while Mersenne Twister is big in size(2 KB).
 - It is less predictable than the MT algorithm since in the case of MT once you have pass 624 numbers you can easily predict the next outputs. 
 
- PCG-64 uses permutation which allows to generate more random numbers.
 
- While Mersenne Twister passes most statistical tests it has experienced some failures while PCG has proven to have very good statistical quality. It can pass Big Crush at 36-bits. 

Although PCG-64 offers so many advantages, it has also some negative qualities such as the fact that it cannot be considered cryptographically secure.
